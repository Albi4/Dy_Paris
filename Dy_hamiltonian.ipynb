```julia
using QuantumOptics
using PyPlot
```

```julia
α=10
β=5
c=0.1
d=100
J=8#Dy gs manifold J=8
```

```julia
# Basis
b = SpinBasis(J)
# Operator creation
Jx = 0.5*sigmax(b)
Jy = 0.5*sigmay(b)
Jz = 0.5*sigmaz(b)
```

```julia
#Hamiltonian
H=-α*Jz-β*Jz^2+c*Jx+d*Jx^2

ψ0=spindown(b)
```

```julia
tspan = [0:0.02:5;]
tout, ψt = timeevolution.schroedinger(tspan, ψ0, H);
```

```julia
using PyPlot

figure(figsize=[10, 3])

subplot(1, 2, 1)
xlabel("Time")
ylabel(L"$ |m_j\rangle $")

plot(tout, real(expect( Jz, ψt)))

```

```julia
energies, states = eigenstates(H);
```

```julia
plot(real(expect(Jz,states)),  energies)
xlabel("m")
ylabel("Energy")
```
